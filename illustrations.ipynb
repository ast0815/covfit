```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Illustration of problem in CDF space
fig, ax = plt.subplots(figsize=(4,4))
ax.set_aspect("equal")
ax.set_xlabel("$y_1 = F_{\chi^2_1}(\Delta^2_1 / S_{11})$")
ax.set_ylabel("$y_2 = F_{\chi^2_1}(\Delta^2_2 / S_{22})$")
ax.set_xlim(0,1)
ax.set_ylim(0,1)
x = 0.3
points = [
    [x, 0],
    [x, x],
    [0, x],
]
ax.plot(*zip(*points), color="k", linestyle="dashed")
ax.plot([0,1], color="k", linestyle="dotted")
ax.annotate(" $x$", (x, x/10))
phi = np.linspace(0, np.pi/2, 50)
cp = np.cos(phi)
sp = np.sin(phi)
r = np.abs(phi - np.pi/4)/2 + np.sqrt(2)*x
ax.fill_between(sp*r, cp*r, hatch="/", facecolor="none", edgecolor="C0")
ax.annotate("$A$", (x/3, x*1.2), fontsize="x-large", color="C0", bbox=dict(boxstyle="round,pad=0.1", facecolor="white", edgecolor="none"))
ax.plot(sp*r, cp*r, color="C1")
ax.annotate("$z = const. = x$", (x*0.8, x*2), color="C1")
fig.tight_layout()
fig.savefig("plots/cdf-space.pdf")
```

```python
# Illustration of invariant1 in CDF space
fig, ax = plt.subplots(figsize=(4,4))
ax.set_aspect("equal")
ax.set_xlabel("$y_1 = F_{\chi^2_1}(\Delta^2_1 / S_{11})$")
ax.set_ylabel("$y_2 = F_{\chi^2_1}(\Delta^2_2 / S_{22})$")
ax.set_xlim(0,1)
ax.set_ylim(0,1)
x = 0.3
points = [
    [x, 0],
    [x, x],
    [0, x],
]
ax.plot(*zip(*points), color="k", linestyle="dashed")
ax.plot([0,1], color="k", linestyle="dotted")
ax.annotate(" $x$", (x, x/10))
points = [
    [0, 1],
    [x, x],
    [1, 0],
]
ax.fill_between(*zip(*points), hatch="/", facecolor="none", edgecolor="C0")
ax.annotate("$A$", (x/3, x*1.2), fontsize="x-large", color="C0", bbox=dict(boxstyle="round,pad=0.1", facecolor="white", edgecolor="none"))
ax.plot(*zip(*points), color="C1")
ax.annotate(r"$z = \dfrac{y_\min}{y_\min + (1 - y_\max)}$", (x*0.5, x*2.3), color="C1")
points = [
    [0.4, 0],
    [1, 0.6],
]
ax.plot(*zip(*points), color="C2")
ax.annotate(r"$\mathbf{y}_s + z(\mathbf{y}_e-\mathbf{y}_s)$", (0.65, 0.2), color="C2")
fig.tight_layout()
fig.savefig("plots/cdf-invariant1.pdf")
```

```python
# Illustration of invariant2 in CDF space
fig, ax = plt.subplots(figsize=(4,4))
ax.set_aspect("equal")
ax.set_xlabel("$y_1 = F_{\chi^2_1}(\Delta^2_1 / S_{11})$")
ax.set_ylabel("$y_2 = F_{\chi^2_1}(\Delta^2_2 / S_{22})$")
ax.set_xlim(0,1)
ax.set_ylim(0,1)
x = 0.3
points = [
    [x, 0],
    [x, x],
    [0, x],
]
ax.plot(*zip(*points), color="k", linestyle="dashed")
ax.plot([0,1], color="k", linestyle="dotted")
ax.annotate(" $x$", (x, x/10))
points = [
    [0, x+(1-x)/2],
    [x, x+(1-x)/2],
    [x, x],
    [x+(1-x)/2, x],
    [x+(1-x)/2, 0],
]
ax.fill_between(*zip(*points), hatch="/", facecolor="none", edgecolor="C0")
ax.annotate("$A$", (x/3, x*1.2), fontsize="x-large", color="C0", bbox=dict(boxstyle="round,pad=0.1", facecolor="white", edgecolor="none"))
ax.plot(*zip(*points), color="C1")
ax.annotate(r"$z = \max(g(y_\max), y_\min)$", (x*0.3, x*2.6), color="C1")
ax.arrow(x*1.1, x*2.5, -x*0.5, -x*0.25, head_width=0.02, color="C1", length_includes_head=True)
ax.arrow(x*1.6, x*2.5, -x*0.5, -x*0.6, head_width=0.02, color="C1", length_includes_head=True)
fig.tight_layout()
fig.savefig("plots/cdf-invariant2.pdf")
```
